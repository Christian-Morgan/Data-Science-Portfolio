```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
sleep_df = pd.read_csv("Sleep_health_and_lifestyle_dataset.csv")
sleep_df.head(10)
```

```python
sleep_df.replace({np.nan:'No_Sleep_Disorder'},inplace=True)
sleep_df
```

```python
sleep_df.replace({'Sleep Apnea':'Sleep_Disorder'},inplace=True)
sleep_df.replace({'Insomnia':'Sleep_Disorder'},inplace=True)
sleep_df
```

```python
sleep_df['Sleep Disorder'] = sleep_df['Sleep Disorder'].astype('category') #Label Encoding
sleep_df['Sleep Disorder'] = sleep_df['Sleep Disorder'].cat.codes
```

```python
sleep_df['Sleep Disorder'] = sleep_df['Sleep Disorder'].astype('category')
sleep_df['Sleep Disorder'] = sleep_df['Sleep Disorder'].cat.codes
```

#### Person ID is not relevant for predicting sleep disorders. Similarly, Occupation is not relevant either because the affects on one's sleep caused by their occupation are already accounted for, e.g., stress level. Therefore, we drop these columns.

```python
sleep_df.drop('Occupation',axis=1,inplace=True)
```

```python
sleep_df.drop('Person ID',axis=1,inplace=True)
```

```python
sleep_df.replace({'Normal Weight':'Normal'},inplace=True)
sleep_df
```

### The following classifications are from https://www.mayoclinic.org/diseases-conditions/high-blood-pressure/in-depth/blood-pressure/art-20050982.

```python
Normal_Blood_P=['118/75', '115/75', '115/78', '118/76', '119/77', '117/76']
Elevated_Blood_P=['121/79']
S1_Hyper=['132/87', '125/82', '130/85', '130/86', '120/80', '125/80', '128/85', '135/88', '126/83', '135/90', '122/80', '128/84', '129/84','131/86']
S2_Hyper=['142/92', '140/90', '139/91', '140/95']
```

```python
replace_dict = {value: 'Normal_Blood_Pressure' for value in Normal_Blood_P}
replace_dict2 = {value: 'Elevated_Blood_Pressure' for value in Elevated_Blood_P}
replace_dict3 = {value: 'Stage_1_Hypertension' for value in S1_Hyper}
replace_dict4 = {value: 'Stage_2_Hypertension' for value in S2_Hyper}
# Replace values in the 'Blood Pressure' column
sleep_df.loc[:, 'Blood Pressure'] = sleep_df['Blood Pressure'].replace(replace_dict)
sleep_df.loc[:, 'Blood Pressure'] = sleep_df['Blood Pressure'].replace(replace_dict2)
sleep_df.loc[:, 'Blood Pressure'] = sleep_df['Blood Pressure'].replace(replace_dict3)
sleep_df.loc[:, 'Blood Pressure'] = sleep_df['Blood Pressure'].replace(replace_dict4)
sleep_df.head(50)
```

```python
sleep_df['Blood Pressure'] = sleep_df['Blood Pressure'].astype('category')
sleep_df['Blood Pressure'] = sleep_df['Blood Pressure'].cat.codes
sleep_df.head(50)
```

```python
sleep_df['BMI Category'] = sleep_df['BMI Category'].astype('category')
sleep_df['BMI Category'] = sleep_df['BMI Category'].cat.codes
sleep_df.head(50)
```

```python
sleep_df['Gender'] = sleep_df['Gender'].astype('category')
sleep_df['Gender'] = sleep_df['Gender'].cat.codes
sleep_df.head(50)
```

```python
sleep_df.corr()
```

```python
X = sleep_df.drop('Sleep Disorder',axis=1)
y = sleep_df['Sleep Disorder']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=21)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler() #creates new StandardScaler object

X_train_scaled = scaler.fit_transform(X_train) 
X_test_scaled = scaler.transform(X_test)
```

```python
log_reg = LogisticRegression(random_state=0).fit(X_train_scaled, y_train)
```

```python
log_reg.score(X_train_scaled, y_train)
```

```python
log_reg.score(X_test_scaled, y_test)
```

```python
# Predict the probabilities for the test set
y_pred_proba = log_reg.predict_proba(X_test_scaled)

# Predict the target values for the test set
y_pred = log_reg.predict(X_test_scaled)

# Evaluate the model using log loss
from sklearn.metrics import log_loss, accuracy_score, confusion_matrix, classification_report

log_loss_value = log_loss(y_test, y_pred_proba)
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

print("Log Loss:", log_loss_value)
print("Accuracy:", accuracy)
print("Confusion Matrix:\n", conf_matrix)
print("Classification Report:\n", class_report)
```
